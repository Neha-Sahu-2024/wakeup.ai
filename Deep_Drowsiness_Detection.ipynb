#### 5. Train from Scratch 

```python
import uuid #unique identifier
import os 
import time
```

```python
Image_Path = os.path.join('data', 'images') # /data/images
labels = ['awake', 'drowsy']
number_imgs =  20
```

```python
cap = cv2.VideoCapture(0)
#loop through labels
for label in labels:
    print('collecting images for {}'.format(label))
    time.sleep(5)

      #  Create label folder if not exists
    label_folder = os.path.join(Image_Path, label)
    if not os.path.exists(label_folder):
        os.makedirs(label_folder)

    
    #loop through image range
    for img_num in range(number_imgs):
        print('collecting images for {}, img number {}'.format(label,img_num))
        ret, frame = cap.read()

        #naming out image path
        imgname = os.path.join(Image_Path, label, f"{uuid.uuid1()}.jpg")

        #write out image to file
        cv2.imwrite(imgname,frame)
        #render to the screen

        cv2.imshow('Image Collection', frame)

        #2second delay between capture
        time.sleep(2)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()                                                                 
```

```python
import os
import shutil
import random
from pathlib import Path

# Set your base image directory
base_dir = Path("C:/Users/Neha/yolov5/data/images")
label_dirs = ['awake', 'drowsy']

# Paths to create
train_dir = base_dir / 'train'
val_dir = base_dir / 'val'

# Create folders if they dont exist
for folder in [train_dir, val_dir]:
    folder.mkdir(parents=True, exist_ok=True)

# Function to move files
def move_files(file_list, target_dir):
    for img_path in file_list:
        txt_path = img_path.with_suffix('.txt')
        shutil.copy(img_path, target_dir / img_path.name)
        if txt_path.exists():
            shutil.copy(txt_path, target_dir / txt_path.name)

# Split each class
for label in label_dirs:
    label_path = base_dir / label
    all_images = list(label_path.glob("*.jpg"))

    random.shuffle(all_images)
    split_index = int(0.8 * len(all_images))
    train_images = all_images[:split_index]
    val_images = all_images[split_index:]

    move_files(train_images, train_dir)
    move_files(val_images, val_dir)

print(" Data has been split into 'train' and 'val' successfully.")
```

```python
import os
import shutil
from pathlib import Path

# Set paths
base_path = Path("C:/Users/Neha/yolov5/data")
image_train = base_path / "images" / "train"
label_train = base_path / "labels" / "train"
image_val = base_path / "images" / "val"
label_val = base_path / "labels" / "val"

# Create label folders if not exist
label_train.mkdir(parents=True, exist_ok=True)
label_val.mkdir(parents=True, exist_ok=True)

# Move .txt files for train
for txt_file in image_train.glob("*.txt"):
    shutil.move(str(txt_file), label_train / txt_file.name)

# Move .txt files for val
for txt_file in image_val.glob("*.txt"):
    shutil.move(str(txt_file), label_val / txt_file.name)

print(" All label .txt files moved to correct folders.")
```
